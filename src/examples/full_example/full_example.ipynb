```python
from pathlib import Path
from sparc import Sparc
```

```python
sam_model_path = "[replace with path to model]/sam_vit_h_4b8939.pth"

root_dir = Path("data")
sol = "1472"
obs_ix = 0

sparc = Sparc(sam_model_path)
sparc.run_pipeline(iof_path=root_dir / f"{sol}/iof", obs_ix=obs_ix)
sparc.export_results(
    f"out/{sol}/{obs_ix}", metadata={"SOL": int(sol)}
)
```

```python
from sparc.visualization.plotting import plot_roi_image

fig = plot_roi_image(
    sparc.load_result["rgb_img"], sparc.final_rois, show=True
)
```
